```python
###Dealing with Outliers

def remove_outliers(df, column):
    Q1 = np.quantile(df[column],0.25)
    Q3 = np.quantile(df[column],0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filtering the data within the bounds
    result = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)] #should satisfy at least one of these conditions
    return result

```

```python

df_clean_glucose = remove_outliers(df, "blood_glucose_level")
df_clean_age = remove_outliers(df, "age")
df_clean_hypertension = remove_outliers(df, "hypertension")
df_clean_heart_disease = remove_outliers(df, "heart_disease")
df_clean_bmi = remove_outliers(df, "bmi")
df_clean_HbA1c_level= remove_outliers(df, "HbA1c_level")
df_clean_diabetes = remove_outliers(df, "diabetes")
```

```python
def redefine_smoking_labels(smoke_status):
    categories = {
        'never': 'non-smoker', 'No Info': 'non-smoker',
        'current': 'current',
        'ever': 'past_smoker', 'former': 'past_smoker', 'not current': 'past_smoker'
    }
    return categories.get(smoke_status, 'unknown')  # default category 'unknown' if no match found

# Map the new function over the column to update the 'smoking_history' entries
df['smoking_history'] = df['smoking_history'].map(redefine_smoking_labels)
```
